# Matplotlib pie chart

```python
import matplotlib.pyplot as plt
```

```python
plt.rc("font", family="Arial Unicode Ms")
```

```python
x=[20,30,50]
total=sum(x)
label=[str(100*data/total)+"%" for data in x]
print(label)
plt.pie([20,30,50], labels=label, labeldistance=0.5)
plt.legend()
plt.title("圓餅圖")
plt.show()
```

```python
import csv
file = open("pie_data.csv", encoding = "utf-8-sig")
reader = csv.reader(file)
next(reader)  #讀取第一列
x=[]
labels=[]
for row in reader:
    labels.append(row[0])
    x.append(int(row[1]))
plt.pie(x, labels=labels, labeldistance=0.5)
plt.legend()
plt.title("程式語言使用率")
plt.show()
```
